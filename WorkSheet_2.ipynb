# Q2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)and make data frame.

```python
url=requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc")
```

```python
soup=BeautifulSoup(url.text, 'html.parser')
soup.title.text
```

```python
movies=soup.find_all('div', class_="lister-item mode-advanced")
```

```python
first_movie=movies[0]
```

```python
#Name of the movie
first_movie.h3.a.text   
```

```python
#Year of the movie
first_movie.find('span', class_="lister-item-year text-muted unbold" ).text[1:5]
```

```python
#Rating of the movie
first_movie.find('div', class_="inline-block ratings-imdb-rating")['data-value']
```

```python
Name = []
Rating= []
Year = []

for i in movies:
    Name.append(i.h3.a.text)
    Year.append(i.find('span', class_="lister-item-year text-muted unbold" ).text[1:5])
    Rating.append(i.find('div', class_="inline-block ratings-imdb-rating")['data-value'])
    
data= list(zip(Name, Year, Rating))
```

```python
import pandas as pd
df=pd.DataFrame(data, columns=['Name', 'Year', 'Rating'])
df.head()
```

```python
Product_Price=[]
price=soup.find_all('span', class_="discountedPriceText")

for i in range(0, 10):
    Product_Price.append(price[i].text)
    
Product_Price
```

```python
Product_image_url=[]
image= soup.find_all('div', class_="productCardImg false")
for i in range(0, 10):
    Product_image_url.append(image[i])
    
Product_image_url
```

```python
data=list(zip(Product_info, Product_Price, Product_image_url))
```

```python
import pandas as pd
df=pd.DataFrame(data, columns=['Product_info', 'Product_Price', 'Product_image_url'])
df
```
